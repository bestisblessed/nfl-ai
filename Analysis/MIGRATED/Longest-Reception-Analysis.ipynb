### 1. **Opponents' Defensive Performance:**
   **How do opposing teams defend against deep passes?**

By identifying defenses that allow the highest percentage of deep passes or YAC (yards after catch), you can exploit WRs who thrive in these conditions.
   
   **Data to Track**:
   - **Opponent's Allowance of 20+ Yard Passes**: Track how often a defense gives up passes of 20+ yards.
   - **Blitzing Frequency**: Teams that blitz frequently may leave one-on-one coverage, increasing the likelihood of long receptions.
   - **Pass Completion % Against Specific Defenses**: Look at completion percentage allowed on deep passes (>20 yards).

   **Edge**:
   - **Bet Overs on WRs against blitz-heavy defenses**: Deep threat WRs (like Tyreek Hill or DeSean Jackson) thrive in one-on-one coverage, which is more common against blitzing defenses.
   - **Example**: If a defense ranks in the bottom 5 in defending deep passes (20+ yards), consider betting the **over** on a WR known for long receptions.

   **Next Step**: 
   - Use tools like PFF (Pro Football Focus) to track which defenses are weak at defending deep passes and which WRs have a strong history of performing well against such defenses.

### 2. **Moving Averages and Trends**:
   **Rolling averages to detect performance trends**:
   - **Analysis**: A rolling average of a player’s longest reception over the last 3–5 games helps to spot trends that the betting market might not have priced in. If a WR's performance is trending upward, there might be value before the line adjusts.
   
   **Data to Track**:
   - **Rolling Averages of Longest Receptions**: Calculate a rolling average (e.g., over the last 3 or 5 games) to smooth out game-by-game variance and reveal underlying trends.
   - **Z-Scores**: Measure how a player’s recent performance compares to their overall average performance using Z-scores to detect anomalies or trends.

   **Edge**:
   - **Bet on improving players before sportsbooks adjust**: If a player’s rolling average is increasing steadily over the last few games, but sportsbooks haven’t raised the line significantly, there’s value in betting the **over**.
   - **Example**: A player like Justin Jefferson might show a pattern of increasing long receptions as the season progresses. If his line remains static while his rolling average rises, there’s an opportunity to exploit.

   **Next Step**:
   - Run a rolling average calculation on WRs' longest receptions over the past 3–5 games, looking for consistent improvement, and bet **overs** before the market catches up.

### 3. **Opponent-Specific Trends**:
   **Performance against specific defenses (man vs. zone, etc.)**:
   - **Analysis**: Certain WRs perform better against specific defensive schemes. For example, some WRs excel against man-to-man coverage due to their speed or ability to beat defenders off the line, while others thrive in zones by finding soft spots.
   
   **Data to Track**:
   - **Defense Scheme Type**: Track whether the opponent uses more man coverage or zone coverage.
   - **WR Performance by Scheme**: Track WR performance specifically against man vs. zone. Some WRs, like Tyreek Hill, thrive in man coverage, while possession WRs like Michael Thomas might excel in zone.

   **Edge**:

   **Next Step**:

### 4. **Exploit Market Overreaction**:
   **Sportsbooks overreact to recent performances**:
   - **Analysis**: After a player has a standout game, sportsbooks may overreact by setting the next game’s line too high. By identifying these overreactions, you can find value in betting the **under** on players who are unlikely to replicate such performances.
   
   **Data to Track**:
   - **Player's Recent Performance vs. Career Averages**: Look for players who recently exceeded their normal performance (e.g., an outlier game with an unusually long reception) but whose career averages suggest they will regress.
   - **Market Line Movement**: Track line movement for longest reception props. If the line moves significantly due to a standout performance, it’s often an overreaction.

   **Edge**:

```python
# Longest Reception in All Career Games

data = pd.read_csv('data/all_passing_rushing_receiving.csv')

# player_name = input("Enter the player's name: ")
# player_name = 'Tyreek Hill'
# player_name = 'Jameson Williams'

def get_player_reception_stats(player_name):
    # Filter the dataset for the given player
    player_data = data[data['player'] == player_name]
    
    # Drop rows with missing rec_long values
    player_data = player_data[['game_id', 'rec_long']].dropna(subset=['rec_long'])
    
    # Group by game_id to avoid duplicates, and take the max rec_long for each game
    player_data = player_data.groupby('game_id').agg({'rec_long': 'max'}).reset_index()
    
    # Sort by game_id to ensure chronological order
    player_data = player_data.sort_values(by='game_id', ascending=False)
    
    # Print game_id and rec_long for each game
    print(f"Longest reception for {player_name} over career (in chronological order):")
    for index, row in player_data.iterrows():
        print(f"Game: {row['game_id']}, Longest Reception: {row['rec_long']}")
    
    # Calculate overall stats
    total_games = len(player_data)
    avg_rec_long = player_data['rec_long'].mean()
    max_rec_long = player_data['rec_long'].max()
    
    # Print overall stats
    print("\n--- Overall Stats ---")
    print(f"Total games: {total_games}")
    print(f"Average longest reception: {avg_rec_long:.2f}")
    print(f"Maximum longest reception: {max_rec_long}")

get_player_reception_stats(player_name)
```

```python
# 3. Double Bar Chart: Last 20 Games vs League Average

def plot_last_10_games_vs_league_avg(player_name):
    # Filter player data and remove missing values
    player_data = data[(data['player'] == player_name) & (data['rec'] > 0)].dropna(subset=['rec_long'])
    
    # Drop duplicates based on game_id to ensure unique games
    player_data = player_data.drop_duplicates(subset=['game_id'])
    
    # Sort by game_id and get the last 10 games
    player_last_10 = player_data.sort_values(by='game_id').tail(20)
    
    # Filter the league data to include only wide receivers who actually played (rec > 0)
    league_data = data[(data['position'] == 'WR') & (data['rec'] > 0) & (data['rec_long'].notna())]
    
    # Calculate league average longest reception for each game week
    league_avg_per_game = league_data.groupby('game_id')['rec_long'].mean().reset_index()
    
    # Filter the league average data for the last 10 games of the player
    league_avg_last_10 = league_avg_per_game[league_avg_per_game['game_id'].isin(player_last_10['game_id'])]
    
    # Plotting the double bar chart
    plt.bar(player_last_10['game_id'], player_last_10['rec_long'], width=0.4, label=player_name, align='center', color='blue')
    plt.bar(league_avg_last_10['game_id'], league_avg_last_10['rec_long'], width=0.4, label='League Average (WRs)', align='edge', color='orange')
    
    plt.title(f'{player_name} vs. League WR Average (Last 10 Games)', fontsize=14)
    plt.xlabel('Game ID', fontsize=12)
    plt.ylabel('Longest Reception (yards)', fontsize=12)
    plt.xticks(rotation=90, fontsize=10)
    plt.legend()
    plt.tight_layout()
    plt.show()

# Example Usage: Tyreek Hill
plot_last_10_games_vs_league_avg(player_name)
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load your dataset (replace 'path_to_your_csv_file.csv' with your actual file path)
# data = pd.read_csv('path_to_your_csv_file.csv')

# Filter player data for Jameson Williams and ensure we only work with games where 'rec_long' is not missing
player_name = 'Jameson Williams'
jameson_data = data[(data['player'] == player_name) & (data['rec_long'].notna())].drop_duplicates(subset=['game_id']).sort_values(by='game_id')

# 1. Add detailed defensive data (mocked for illustration, replace with real data)
defensive_details = {
    'LA': {'deep_pass_completion_rate': 0.45, 'pass_rush_efficiency': 80, 'cb_grade': 65},
    'TB': {'deep_pass_completion_rate': 0.38, 'pass_rush_efficiency': 78, 'cb_grade': 75},
    # Add more teams here
}

# Add defensive details to the dataset
jameson_data['deep_pass_completion_rate'] = jameson_data['opponent_team'].map(lambda x: defensive_details.get(x, {}).get('deep_pass_completion_rate', None))
jameson_data['pass_rush_efficiency'] = jameson_data['opponent_team'].map(lambda x: defensive_details.get(x, {}).get('pass_rush_efficiency', None))
jameson_data['cb_grade'] = jameson_data['opponent_team'].map(lambda x: defensive_details.get(x, {}).get('cb_grade', None))

# 2. Calculate Z-Scores for Longest Reception
mean_rec_long = jameson_data['rec_long'].mean()
std_rec_long = jameson_data['rec_long'].std()

# Add Z-score column
jameson_data['z_score'] = (jameson_data['rec_long'] - mean_rec_long) / std_rec_long

# 3. Plot Rolling Averages and Z-Scores for deeper analysis
def plot_rolling_averages_and_z_scores(jameson_data):
    plt.figure(figsize=(10, 6))
    plt.plot(jameson_data['game_id'], jameson_data['rec_long'], marker='o', label='Actual Longest Reception')
    plt.plot(jameson_data['game_id'], jameson_data['z_score'], linestyle='--', color='orange', label='Z-Score')
    plt.title(f"Jameson Williams: Longest Reception and Z-Scores")
    plt.xlabel('Game ID')
    plt.ylabel('Yards / Z-Score')
    plt.xticks(rotation=90)
    plt.grid(True)
    plt.legend()
    plt.tight_layout()
    plt.show()

# Call the function to visualize rolling averages and Z-Scores
plot_rolling_averages_and_z_scores(jameson_data)

# 4. Output the data for review (useful for manual inspection)
print(jameson_data[['game_id', 'rec_long', 'z_score', 'deep_pass_completion_rate', 'pass_rush_efficiency', 'cb_grade']])

# 5. Summary of key findings and betting insights
def summarize_insights(jameson_data):
    last_game = jameson_data.tail(1)
    z_score = last_game['z_score'].values[0]
    deep_pass_rate = last_game['deep_pass_completion_rate'].values[0]
    pass_rush_efficiency = last_game['pass_rush_efficiency'].values[0]
    cb_grade = last_game['cb_grade'].values[0]

    print(f"\n### Summary of Insights for Jameson Williams ###")
    print(f"Last Game's Longest Reception: {last_game['rec_long'].values[0]} yards")
    print(f"Z-Score: {z_score:.2f}")
    
    if deep_pass_rate:
        print(f"Opponent's Deep Pass Completion Rate: {deep_pass_rate * 100}%")
    if pass_rush_efficiency:
        print(f"Opponent's Pass Rush Efficiency: {pass_rush_efficiency}")
    if cb_grade:
        print(f"Opponent's Cornerback Grade: {cb_grade}")

    if z_score > 2:
        print("Performance: Jameson Williams had an unusually high performance. Be cautious of market overreaction in the next game.")
    else:
        print("Performance: Williams' performance is within his typical range.")

    if deep_pass_rate > 0.40 and cb_grade < 70:
        print("Opportunity: The opponent's secondary is vulnerable to deep passes. Consider betting the over.")
    else:
        print("Consideration: Opponent's secondary may limit deep passes. Be cautious with overs.")

# Call the summary function for the last game
summarize_insights(jameson_data)

# Export the processed data to a CSV for future reference (optional)
# jameson_data.to_csv('jameson_williams_deeper_analysis.csv', index=False)
```
